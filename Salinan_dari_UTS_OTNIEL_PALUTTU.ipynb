```python
import requests
from bs4 import BeautifulSoup

List_Buku = []

for page in range(1, 5):
    url = f"https://deepublishstore.com/product-category/ekonomi/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    data = soup.find_all("li", class_="product")

    for item in data:
        # Judul
        judul = item.find("h2", class_="woocommerce-loop-product__title").text.strip()

        # Harga
        harga = item.find("span", class_="woocommerce-Price-amount amount").text.strip()

        # Gambar
        gambar = item.find("img", class_="attachment-woocommerce_thumbnail")["src"]

        List_Buku.append({
            "judul": judul,
            "harga": harga,
            "gambar": gambar
        })

# cek hasil
List_Buku[:5]
```

```python
import json

with open("List_Buku.json", "w", encoding="utf-8") as f:
    json.dump(List_Buku, f, indent=4, ensure_ascii=False)
```

```python
import pandas as pd

# Extract
df = pd.read_json("List_Buku.json")

# Transform (bersihkan harga)
df["harga"] = df["harga"].str.replace("Rp", "")
df["harga"] = df["harga"].str.replace(".", "")
df["harga"] = df["harga"].str.strip()
df["harga"] = df["harga"].astype(int)

# Load ke CSV
df.to_csv("List_Buku.csv", index=False)

df.head()
```
